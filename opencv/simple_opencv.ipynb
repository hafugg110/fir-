```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = cv2.imread("cat.jpg")
```

```python
plt.imshow(f)
```

**顏色怪怪的**

opencv 的讀圖預設以 BGR 的方式讀入，所有讀入的圖片必須要轉換成 RGB


```python
f_rgb = cv2.cvtColor(f, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(f_rgb)
```

```python
# 查看圖片尺寸(長/寬/Channel)
f_rgb.shape
```

```python
# 檢查圖片型態
type(f_rgb)
```

```python
# 將 Channel反轉可以得到一樣的顏色效果
plt.imshow(f[:,:,::-1])
```

```python
# 只各取出 1/2 長寬的圖片（全圖1/4）
plt.imshow(f[800:,700:,::-1])
```

```python
# 改變圖片尺寸
f_rgb_resize = cv2.resize(f_rgb, (int(1200*0.5), int(675*0.5)), cv2.INTER_AREA) 
print(f_rgb_resize.shape)
plt.imshow(f_rgb_resize)
```

```python
# 疊合兩張圖片（兩圖需相同尺寸）
## 函式寫法
# cv2.addWeighted(圖片1, 比例, 圖片2, 比例, gamma= 調亮度)
##
dst = cv2.addWeighted(f_rgb, 0.8, f, 0.2, gamma= 100)
plt.imshow(dst)
```

```python
# 儲存圖片檔案
cv2.imwrite("2cat.jpg", dst)
```
