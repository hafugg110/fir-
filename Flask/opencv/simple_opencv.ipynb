```python
# 
f_rgb_resize = cv2.resize(f_rgb, (int(1200*0.5), int(675*0.5)), cv2.INTER_AREA) 
print(f_rgb_resize.shape)
plt.imshow(f_rgb_resize)
```

```python
# 
## 
# cv2.addWeighted(1, , 2, , gamma= )
##
dst = cv2.addWeighted(f_rgb, 0.8, f, 0.2, gamma= 100)
plt.imshow(dst)
```

```python
# 
cv2.imwrite("2cat.jpg", dst)
```
